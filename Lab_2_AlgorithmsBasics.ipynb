```python
# Exercise 1
def func1(a, b):
  result = 0
  for i in range(0, b):
    result += a
  return result
```

```python
# Exercise 2
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a
    b = b - 1
  return result
```

```python
import time
start = time.time()
print(func2(4,3))
end=time.time()
ellapsed =end-start
print(ellapsed)
#THE FUNCTION RETURNS a RAISED TO b HENCE PERFORMING THE POWER FUNCTION
```

```python
# Exercise 3
def func3(num):
   result = 0
   while num > 0:
    result = result + num % 10
    num //= 10
    return result
```

```python
# Sample function call: func3(1234)
import time
start = time.time()
print(func3(43))
end=time.time()
ellapsed =end-start
print(ellapsed)
#THE FUNCTION RETURNS THE LAST DIGIT OF THE NUMBER
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
import time
start = time.time()
print(func4([1, 2, 3, 4], [2, 6, 6, 8]))
end=time.time()
ellapsed =end-start
print(ellapsed)
#THE FUNCTION SHOWS THE NO OF INTERSECTING ELEMENTS IN BOTH THE LISTS USING BINARY SEARCH
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
import time
start = time.time()
print(func5(43))
end=time.time()
ellapsed =end-start
print(ellapsed)
#THE FUNCTION GUESSING GAME IT RETURNS THE GUESS VALUE WHEN THE SQUARE OF THE GUESSED VALUE IS CORRECTLY GUESSED AND RETURNS NONE WHEN NOT GUESSED
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
#Sample function call: func8([3,1,2,7,8,6])
import time
start = time.time()
print(func8([3,1,2,7,8,6]))
end=time.time()
ellapsed =end-start
print(ellapsed)
#THE FUNCION PERFORMS SWAPING OF THE ELEMENTS IN THE ARRAY
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
import time
start = time.time()
print(func12([2,4,5,3],2))
end=time.time()
ellapsed =end-start
print(ellapsed)
#THE FUNCTION PERFORMS THE SWAPPING OF THE DIVIDED ARRAYS
```

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(a):
    n=len(a)
    sum=0
    for i in range(n):
      sum+=a[i]
    return sum
```

```python
#call the function
arr = [12, 3, 4, 15,45,78,450]
import time
start = time.time()
print(sum_arr(arr))
end=time.time()
ellapsed =end-start
print(ellapsed)
#THE FUNCTION RETURNS THE SUM OF ALL THE ELEMENTS OF THE ARRAY
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(a,k):
  n=len(a)
  if (k==1):
    max=a[0]
    for i in range(n):
      if (max<a[i]): max=a[i]
    return max
  min=a[0]
  for i in range(n):
    if(min>a[i]): min=a[i]
  return min
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(a):
  n=len(a)
  e,o=0,0
  for i in range(n):
    if(a[i]%2==0): e+=1
    if(a[i]%2==1): o+=1
  print("even=",e)
  print("odd=",o)
```
